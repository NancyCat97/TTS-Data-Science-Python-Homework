### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
item_counts = chipo['item_name'].value_counts()
top_items = item_counts.head(5)
plt.bar(top_items.index, top_items.values)
plt.xlabel('Items')
plt.ylabel('Purchase Count')
plt.title('Top 5 Items Bought at Chipotle')
plt.xticks(rotation=45)
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
# create a list of prices
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] # strip the dollar sign and trailing space
# groupby the orders and sum
orders = chipo.groupby('order_id').sum(numeric_only=True)

# creates the scatterplot
plt.scatter(x = orders.item_price, y = orders.quantity, s = 50, c = 'black')

# Set the title and labels
plt.xlabel('Items Ordered')
plt.ylabel('Price')
plt.title('Number of items ordered by order price')
plt.ylim(0)
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# Top 10 most popular items ordered
item_counts = chipo['item_name'].value_counts()
top_items = item_counts.head(10)
plt.bar(top_items.index, top_items.values)
plt.xlabel('Items')
plt.ylabel('Price')
plt.title('Top 10 Most Popular Items Bought at Chipotle')
plt.xticks(rotation=45)
plt.show()
```
